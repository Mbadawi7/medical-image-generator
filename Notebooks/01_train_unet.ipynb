```python
import torch
from monai.networks.nets import UNet
from monai.losses import DiceLoss

# Model
model = UNet(
    spatial_dims=2,
    in_channels=1,
    out_channels=1,
    channels=(16, 32, 64, 128, 256),
    strides=(2, 2, 2, 2),
).to("cpu")

loss_function = DiceLoss(sigmoid=True)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

print("Model ready ✅")
```
